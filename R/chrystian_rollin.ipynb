```r
##Se Lee el archivo Calendario.xlsx
df_calendario <- read_excel("/content/Calendario.xlsx", sheet = "Calendario")
df_calendario$Fecha <- as.Date(df_calendario$Fecha)

#Se realiza una transformación para traducir de ingles a español el nombre de los meses
meses <- c(
"enero","febrero","marzo","abril","mayo","junio",
"julio","agosto","septiembre","octubre","noviembre","diciembre"
)

df_calendario$Numero_Mes <- as.numeric(format(df_calendario$Fecha,"%m"))
df_calendario$Nombre_Mes <- meses[df_calendario$Numero_Mes]
df_calendario$Anio <- format(df_calendario$Fecha,"%Y")
print(paste("Cantidad de registros de Calendario:", nrow(df_calendario)))
head(df_calendario)
```

```r
##Se Lee el archivo Cliente.xlsx
df_Clientes1 <- read_excel("/content/Cliente.xlsx", sheet = "Clientes1")
df_Clientes2 <- read_excel("/content/Cliente.xlsx", sheet = "Clientes2")
df_Clientes3 <- read_excel("/content/Cliente.xlsx", sheet = "Clientes3")

##Se realiza el join por IdCliente estre los data frame
df_clientes_tmp1 <- merge(df_Clientes1, df_Clientes2, by = "IdCliente")
df_clientes <- merge(df_clientes_tmp1, df_Clientes3, by = "IdCliente")

print(paste("Cantidad de registros de clientes:", nrow(df_clientes)))
head(df_clientes)
```

```r
##Se Lee el archivo Territorio.csv
df_territorio <- read.table(
  "/content/Territorio.csv",
  header = TRUE,
  sep = ";",
  fill = TRUE,
  quote = "\"",
  fileEncoding = "latin1"
)

print(paste("Cantidad de registros de Territorio:", nrow(df_territorio)))
head(df_territorio)
```

```r
##Se Lee el archivo Vendedor.xlsx
df_Vendedor <- read_excel("/content/Vendedor.xlsx", sheet = "Vendedor")
print(paste("Cantidad de registros Total de Vendedor:", nrow(df_Vendedor)))
head(df_Vendedor)
```

```r
##Se Lee el archivo Productos.xlsx
df_Productos <- read_excel("/content/Productos.xlsx", sheet = "Prod")
print(paste("Cantidad de registros Total de Productos:", nrow(df_Productos)))
head(df_Productos)
```

```r
##Se Leen los archivos de 2017.xlsx, 2018.xlsx y 2019.xlsx
df_data_2019 <- read_excel("/content/Data/2019.xlsx", sheet = "Data")
df_data_2018 <- read_excel("/content/Data/2018.xlsx", sheet = "Data")
df_data_2017 <- read_excel("/content/Data/2017.xlsx", sheet = "Data")

df_data <- rbind(df_data_2017, df_data_2018, df_data_2019)
df_data$Fecha <- as.Date(df_data$Fecha)
print(paste("Cantidad de registros Total de 3 años:", nrow(df_data)))
head(df_data)
```

```r
##exportar data frames a un excel
library(openxlsx)
write.xlsx(df_calendario, "/content/df_calendario.xlsx")
write.xlsx(df_clientes, "/content/df_clientes.xlsx")
write.xlsx(df_territorio, "/content/df_territorio.xlsx")
write.xlsx(df_Vendedor, "/content/df_Vendedor.xlsx")
write.xlsx(df_Productos, "/content/df_Productos.xlsx")
write.xlsx(df_data, "/content/df_data.xlsx")
#lista <- list(
#  Calendario = df_calendario,
#  Clientes = df_clientes,
#  Territorio = df_territorio,
#  Vendedor = df_Vendedor,
#  Productos = df_Productos,
#  Data = df_data
#)
#
#write.xlsx(lista, "/content/resultado.xlsx")
```
